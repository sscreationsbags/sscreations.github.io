```python
import pandas as pd
import string
```

```python
df = pd.read_excel("sscreations_data_updated.xlsx")
df
```

```python
df = df.sort_values(by=['category'])
```

```python
df['count'] = 0
df['category_joined'] = 0
```

```python
df2 = df.groupby('category')['count'].count()
count_mapping = df2.to_dict()
print(count_mapping)
```

```python
for index, row in df.iterrows():
    df.loc[index,['count']] = count_mapping[str(row['category'])]
    df.loc[index,['category_joined']] = str(row['category']).translate({ord(c): None for c in string.whitespace}).lower()
df = df.sort_values(by=['count'], ascending=False)
df
```

```python
prev_category = ""
all_category_joined = []
i = 0
for index, row in df.iterrows():
    category = str(row["category"])
    category_joined = str(row["category_joined"])
    
    if category_joined not in all_category_joined:
        all_category_joined.append(category_joined)
        
    if category != prev_category:
        if i != 0:
            print('</div>')
            # print('<br><br>')
        print('<div class="cards-category-heading" data-item="' + category_joined + '">' + category + '</div>')
        print('<div class="container">')
        prev_category = category
        i += 1
    image_path = str(row["image_path"])
    printing = str(row["printing"])
    description = str(row['description'])
    name = str(row['name'])
    if image_path == "nan":
        image_path = ""
    if printing == "nan":
        printing = ""
    if description == "nan":
        description = ""
    print('<div class="card" data-item="' + category_joined + '"><div class="img"><img src=assets/pictures/bags/' + image_path + '></div><div class="card-text"><h2>' + name + '</h2><h3>Printing Style: ' + printing + '</h3><h3>Dimensions: ' + description + '</h3></div></div>')
print('</div>')
```
